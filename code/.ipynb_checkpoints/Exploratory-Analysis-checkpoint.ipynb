```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, when, isnan, split
import pandas as pd
import numpy as np
from pyspark.ml.feature import StringIndexer
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.stat import Correlation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spSession = SparkSession.builder.master("local").appName("Local-Session").getOrCreate()
```

### Loading train and test data

```python
trainDF1 = spSession.read.csv("../dataset/train.csv", header = True, inferSchema = True).drop("_c0") 
testDF1 = spSession.read.csv("../dataset/test.csv", header = True, inferSchema = True).drop("_c0")  
type(trainDF1)
```

### Looking for NA values

```python
nan = trainDF1.select([count(when(isnan(c), c)) for c in trainDF1.columns]).toPandas()
print("Number of NA values on train data: " + str(int(nan.sum(axis = 1))))
nan = testDF1.select([count(when(isnan(c), c)) for c in testDF1.columns]).toPandas()
print("Number of NA values on test data: " + str(int(nan.sum(axis = 1))))
```

### Checking balance according to target variable

```python
trainDF1.groupBy(trainDF1.churn == "yes").count().show()
```

### Balacing train dataset

```python
trainDFPos = trainDF1.filter(col("churn") == "yes")
trainDFNeg = trainDF1.filter(col("churn") == "no").sample(False, 483/2850)
trainDF2 = trainDFPos.union(trainDFNeg)
trainDF2.groupBy(col("churn") == "yes").count().show()
testDF2 = testDF1
print("Total: " + str(trainDF2.count()))
```

### Turning string columns into numeric

```python
str_columns = [item[0] for item in trainDF2.dtypes if item[1].startswith('string')]
trainDF3 = trainDF2
testDF3 = testDF2

for c in str_columns:
    indexer = StringIndexer(inputCol = c, outputCol = str(c + "_indexed"))
    trainDF3 = indexer.fit(trainDF2).transform(trainDF3)
    testDF3 = indexer.fit(trainDF2).transform(testDF3)
    trainDF3 = trainDF3.drop(c)
    testDF3 = testDF3.drop(c)
```

### Analysing features

```python
vecAssembler = VectorAssembler(inputCols=trainDF3.columns, outputCol="features")
train_vector = vecAssembler.transform(trainDF3)
corr = Correlation.corr(train_vector, "features")
corr = corr.collect()[0]["pearson({})".format("features")].values.reshape((20,20))
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax, xticklabels= trainDF3.columns, yticklabels= trainDF3.columns)
```

### Removing variables with high correlation between each other

```python
remove = ["voice_mail_plan_indexed", "total_day_minutes", 
          "total_eve_minutes", "total_night_minutes", "total_intl_minutes"]
```

```python
trainDF4 = trainDF3
testDF4 = testDF3

for i in remove:
    trainDF4 = trainDF4.drop(i)
    testDF4 = testDF4.drop(i)
```

```python
vecAssembler = VectorAssembler(inputCols=trainDF4.columns, outputCol="features")
train_vector = vecAssembler.transform(trainDF4)
corr = Correlation.corr(train_vector, "features")
corr = corr.collect()[0]["pearson({})".format("features")].values.reshape((15,15))
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax, xticklabels= trainDF4.columns, yticklabels= trainDF4.columns)
```

### Analysing correlation to churn

```python
for c in trainDF4.columns:
    print(c + ": " + str(abs(round(trainDF4.corr("churn_indexed", c, method = "pearson"), 2))) + "\n")
```

### Removing features with low correlation to churn

```python
trainDF5 = trainDF4[[c for c in trainDF4.columns if 
                     abs(trainDF4.corr("churn_indexed", c, method = 'pearson')) > 0.01]] 
testDF5 = testDF4[trainDF5.columns]
```

```python
trainDF5.toPandas()
```

### Extracting more features

```python
trainDF5.toPandas().groupby("churn_indexed").total_day_charge.plot(kind='kde', title = "total_day_charge")
```

```python
# Creating a new variable

trainDF5 = trainDF5.withColumn("total_day_charge_level", when(col("total_day_charge") < 38, 0).otherwise(1))
testDF5 = testDF5.withColumn("total_day_charge_level", when(col("total_day_charge") < 38, 0).otherwise(1))
trainDF5.toPandas().groupby("churn_indexed").total_day_charge_level.plot(kind='kde', 
                                                                         title = "total_day_charge_level")
```

```python
trainDF5.toPandas().groupby("churn_indexed").number_customer_service_calls.plot(
    kind='kde', title = "number_customer_service_calls")
```

```python
# Creating a new variable

trainDF5 = trainDF5.withColumn("number_customer_service_calls_level", 
                               when(col("number_customer_service_calls") < 3, 0).otherwise(1))
testDF5 = testDF5.withColumn("number_customer_service_calls_level", 
                               when(col("number_customer_service_calls") < 3, 0).otherwise(1))
trainDF5.toPandas().groupby("churn_indexed").number_customer_service_calls_level.plot(kind='kde', 
                                        title = "total_day_charge_levelnumber_customer_service_calls_level")
```

```python
for c in trainDF5.columns:
    print(c + ": " + str(abs(round(trainDF5.corr("churn_indexed", c, method = "pearson"), 2))) + "\n")
```
